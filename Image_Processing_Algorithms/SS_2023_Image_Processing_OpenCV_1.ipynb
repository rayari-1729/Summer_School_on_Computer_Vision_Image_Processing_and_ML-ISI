**Image Binarization**

```python
img_catdog = cv2.imread('cat dog.jpg')
plt.imshow(img_catdog)
img_cd_gray = cv2.cvtColor(img_catdog, cv2.COLOR_BGR2GRAY)
plt.imshow(img_cd_gray, cmap = 'gray')
```

```python
ret, thr_img = cv2.threshold(img_cd_gray,127,255,cv2.THRESH_BINARY)
# ret, thr_img = cv2.threshold(img_cd_gray,0,255,cv2.THRESH_OTSU)
print(ret)
plt.imshow(thr_img, cmap = 'gray')
print(thr_img.shape)
```

https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

**Morphological Operation 1**

```python
thr_img2 = np.copy(thr_img)
kernel_size = 15
h_sz = int(kernel_size/2)
kernel = np.ones(kernel_size)

for i in range(h_sz,thr_img.shape[0]-h_sz):
  for j in range(h_sz,thr_img.shape[1]-h_sz):

    tmp = thr_img[i-h_sz : i+h_sz, j-h_sz : j+h_sz]
    max_val = np.amax(tmp)
    thr_img2[i,j] = max_val

# plt.imshow(thr_img2, cmap = 'gray')
f, axarr = plt.subplots(1,2)
axarr[0].imshow(thr_img, cmap = 'gray')
axarr[0].title.set_text('Original image')
axarr[1].imshow(thr_img2, cmap = 'gray')
axarr[1].title.set_text('After Morphological Operation')
```

Kernel.png

**Dilation : This operation widens the boundary of the foreground object**

```python
thr_img2 = np.copy(thr_img)
kernel_size = 15
kernel = np.ones((kernel_size,kernel_size), np.uint8)
img_dilation = cv2.dilate(thr_img2, kernel, iterations=1)
# plt.imshow(img_dilation, cmap = 'gray')

f, axarr = plt.subplots(1,2)
axarr[0].imshow(thr_img, cmap = 'gray')
axarr[0].title.set_text('Original image')
axarr[1].imshow(img_dilation, cmap = 'gray')
axarr[1].title.set_text('After Dilation')
```

**Morphological Operation 2**

```python
thr_img2 = np.copy(thr_img)
kernel_size = 11
h_sz = int(kernel_size/2)
kernel = np.ones(kernel_size)

for i in range(h_sz,thr_img.shape[0]-h_sz):
  for j in range(h_sz,thr_img.shape[1]-h_sz):

    tmp = thr_img[i-h_sz : i+h_sz, j-h_sz : j+h_sz]
    min_val = np.amin(tmp)
    thr_img2[i-h_sz : i+h_sz, j-h_sz : j+h_sz] = min_val

# plt.imshow(thr_img2, cmap = 'gray')
f, axarr = plt.subplots(1,2)
axarr[0].imshow(thr_img, cmap = 'gray')
axarr[0].title.set_text('Original image')
axarr[1].imshow(thr_img2, cmap = 'gray')
axarr[1].title.set_text('After Morphological Operation')
```

**Erosion : This operation destroys the boundary of the foreground object.**

```python
thr_img2 = np.copy(thr_img)
kernel_size = 11
kernel = np.ones((kernel_size,kernel_size), np.uint8)
img_erosion = cv2.erode(thr_img2, kernel, iterations=1)
# plt.imshow(img_erosion, cmap = 'gray')

f, axarr = plt.subplots(1,2)
axarr[0].imshow(thr_img, cmap = 'gray')
axarr[0].title.set_text('Original image')
axarr[1].imshow(thr_img2, cmap = 'gray')
axarr[1].title.set_text('After Erosion')
```
