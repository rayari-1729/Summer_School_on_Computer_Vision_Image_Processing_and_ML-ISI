
Python Imaging Library is a free and open-source additional library for the Python programming language that adds support for opening, manipulating, and saving many different image file formats. It is available for Windows, Mac OS X and Linux. 
Python Pillow is built on the top of PIL (Python Image Library). 



**Opening and saving an image**

This code is for displaying an image and then saving the image. 

The Image module provides a class with the same name which is used to represent a PIL image. 

```python
# Import Image class from the library
from PIL import Image

# Read the image from its location
img = Image.open("imagekolkata.jpg") 
 
# For showing the image
img.show()
display(img) # This is required for colab

# save an image using extension
img1 = img.save("imagekolkata2.jpg") 
```

**Changing the format**

```python
# Read the image from its location and display
img = Image.open("imagekolkata.jpg") 
display(img) # This is required for colab

# format of the image
print(img.format)

# changing to png
img.save("imagekolkata2.png")

imgpng = Image.open("imagekolkata2.png") 
display(imgpng)
print(imgpng.format)

```

**Adding a simple text to an image**

The ImageDraw module provides simple 2D graphics for Image objects. 

Syntax: obj.text( (x,y), Text, fill)

Parameters:

(x, y): This X and Y denotes the starting position(in pixels)/coordinate of adding the text on an image.

Text: A Text or message that we want to add to the Image.

Fill: Fill is to give the Font color to the text.

```python
from PIL import ImageDraw
 
# Call draw Method to add 2D graphics in an image
imgt = ImageDraw.Draw(img)

# Add Text to an image 
imgt.text((20,10), "Howrah Bridge", fill=(0,0, 255))

# Display and save the image with text
display(img)
img.save("howrah bridge.jpg")
```

**Changing the mode of the image**

An RGB color value is specified with: rgb(red, green, blue). Each parameter (red, green, and blue) defines the intensity of the color as an integer between 0 and 255.

Mode of the original image is displayed. Then the 
RGB image is changed to black & white and grayscale images.
Mode descriptions
1 1-bit pixels, black and white
L 8-bit pixels, grayscale 
RGB 3×8-bit pixels, true colour

```python
from PIL import ImageOps

# mode of the original image
#img = Image.open("imagekolkata.jpg") 
display(img)
print(img.mode)

#convert to binary 
imbw = img.convert("1")
display(imbw)
print(imbw.mode) # format of the image

# save the binary(black and white) image 
bw = imbw.save("bwimagekolkata.jpg") 

#convert to binary without dithering
imbw1 = img.convert('1',dither=Image.NONE)
display(imbw1)
print(imbw1.mode) # format of the image

# save the binary(black and white) image 
bw1 = imbw1.save("bwimagekolkata1.jpg") 

# convert to grayscale 
imgray1 = img.convert("L")
display(imgray1)
print(imgray1.mode) # format of the image

imgray = ImageOps.grayscale(img)
display(imgray)
print(imgray.mode) # format of the image
# save the grayscale image using extension
gray = imgray.save("grayimagekolkata.jpg") 
```

**Finding size of the image and then resizing the image**

The size of the original image is displayed. Then a resized copy of the image is returned.

Interpolation (Nearest-neighbor interpolation also known as zero-order interpolation is used here) happens during the resize process, due to which the quality of image changes whether it is being upscaled (resized to a higher dimension than original) or downscaled (resized to a lower Image then original). 


```python
# size of the image
print("Size of original image is",img.size)
img.show()
display(img)

# fetching the original dimensions
wid, hgt = img.size
  
# displaying the original dimensions
print("Dimensions of the original image is",str(wid) + "x" + str(hgt))
#print("Dimensions of the original image is",str(hgt) + "x" + str(wid))

newsize = (550, 200)
imgr = img.resize(newsize)
 
# Shows the image in image viewer
imgr.show()
display(imgr)
print("Size of resized image is",imgr.size)
```

**Cropping an image**



```python
print("Size of original image is",img.size)
img.show()
display(img)
# Setting the points for cropped image
left = 40
top = 100
right = 450
bottom = 200
 
# Cropped image of above dimension
# (It will not change original image)
imgc = img.crop((left, top, right, bottom))
display(imgc)
print("Size of cropped image is",imgc.size)
```

**Rotating the image**



```python
#rotating an image 
imrot = img.rotate(90,expand=0)
# imrot = img.rotate(170, Image.NEAREST, expand = 1)
display(imrot)
imgr = imrot.save("imagekolkatarot.jpg") 
print("170 deg rotation of the image")
```

**Flipping an image**

```python
#vertical flip
vertimg = img.transpose(method=Image.FLIP_TOP_BOTTOM)
display(vertimg)
vertimg.save("vflipkolkata.jpg")
print("Vertical flip of the image") 

#horizontal flip
horimg = img.transpose(method=Image.FLIP_LEFT_RIGHT)
display(horimg)
horimg.save("hflipkolkata.jpg")
print("Horizontal flip of the image") 
```

uint8 represents Unsigned integer (0 to 255)

Image.fromarray -> returns an image object

```python
import numpy as np

print(type(img))
# convert image to numpy array
arrimg= np.array(img)    
#arrimg= np.uint8(img)    #can also use this

print(type(arrimg))

# summarize shape
print(arrimg.shape)

#Showing value of each pixel of the NumPy array image
print(arrimg)
display(arrimg) #shows the data type as well

print('R channel is')
print(arrimg[:,:,0])

print('G channel is')
print(arrimg[:,:,1])
print('B channel is')
print(arrimg[:,:,2])

#using asarray function
from numpy import asarray

data = asarray(img)
print(type(data))


```

**Creating Pillow image from array**



```python
#Convert array to PIL image
PILimg = Image.fromarray(np.array(img))
#PILimg = Image.fromarray(arrimg)

display(PILimg)
print(type(PILimg))

PILimg.save("imagekolkatapil.jpg") #save image

# summarize image details
print(PILimg.mode)
print(PILimg.size)

```

**Converting Images to Grayscale, Resizing and Trimming in numpy**

```python
# Converting to grayscale
im = np.array(img.convert('L')) 

display(im) 

display(Image.fromarray(im))  #PIL image
Image.fromarray(im).save('gr_img.png')

# Resizing
img_rz = np.array(img.resize((340,100)))
print("After resizing:",img_rz.shape)

display(Image.fromarray(img_rz))   # PIL image
Image.fromarray(img_rz).save('r_img.jpeg')

#Trimming the Image
im1 = np.array(img)
print("Before trimming:",im1.shape)

im_trim = im1[50:200, 100:350]
print("After trimming:",im_trim.shape)

display(Image.fromarray(im_trim))   # PIL image
Image.fromarray(im_trim).save('trim_img.jpeg')
```
